```python
#performance of mathematical operation on the given datasets multiplication,division,power off,subtraction
import numpy as np
dataset1=np.array([4,8,12,24])
dataset2=np.array([0.1,0.01,0.001,0.0001])
product=np.multiply(dataset1,dataset2)
product1=dataset1*dataset2
print("product is as follows",product)
print("another product :",product1)
Division=np.divide(dataset1,dataset2)
Moddivision=np.mod(dataset1,dataset2)
Floordivision=np.floor_divide(dataset1,dataset2)
Subtraction=np.subtract(dataset1,dataset2)
print("div",Division)
print("modulp",Moddivision)
print("floor div",Floordivision)
print("sub is",Subtraction)
```

```python
#implemnt trigonmetric and log functions of a given data set
import numpy as np
patient_age=np.array([65,78,45,67,56,77])
sin1=np.sin(patient_age)
cosine=np.cos(patient_age)
tangent=np.tan(patient_age)
log1=np.log10(patient_age)
log=np.log2(patient_age)
exp=np.exp(patient_age)
print("sine of patient age:",sin1)
print("cosine of patient age:",cosine)
print("tangent of patient age:",tangent)
print("logarithmic of patient age:",log1)
print("logarithmic  base 2 of patient age:",log)
print("antilogarithmic result of patient age:",exp)
```
